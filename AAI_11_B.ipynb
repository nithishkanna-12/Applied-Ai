```python

cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
from google.colab import files
uploaded = files.upload()


image_path = list(uploaded.keys())[0]


img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
objects = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

print("Detected objects:", len(objects))
```

```python
output = img_rgb.copy()

for (x, y, w, h) in objects:
    cv2.rectangle(output, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(output)
plt.title("Detected Objects")
plt.axis('off')
```

```python
count = len(objects)
print("Total Objects Count:", count)
```

```python
measurements = []

for i, (x, y, w, h) in enumerate(objects):
    area = w * h
    measurements.append((i+1, w, h, area))
    print(f"Object {i+1}: Width={w}, Height={h}, Area={area}")
```

```python
for i, (x, y, w, h) in enumerate(objects):
    crop = img_rgb[y:y+h, x:x+w]

    plt.imshow(crop)
    plt.title(f"Object {i+1}")
    plt.axis('off')
    plt.show()
```
